## 1.1 Loading Data

```python
import pandas
data = pandas.read_csv('/content/drive/My Drive/Colab Notebooks/Assignment 9/preprocessed_data.csv',nrows=50000)
data.head()
```

```python
y = data['project_is_approved'].values
x = data.drop(['project_is_approved'],axis=1)
print(x.shape)
print(y.shape)
```

```python
#dividing dataset in train,test and validation
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=1)
```

```python
print("-"*50)
print("shape of x_train is {}".format(x_train.shape))
print("shape of x_test is {}".format(x_test.shape))
```

```python
#encoding categorical features 
from sklearn.feature_extraction.text import CountVectorizer

#school state
categ_model = CountVectorizer()
categ_model.fit(x_train['school_state'].values)
school_state_tr = categ_model.transform(x_train['school_state'].values)
school_state_tst = categ_model.transform(x_test['school_state'].values)

print("-"*50)
print("shape of school state (train) vector is {}".format(school_state_tr.shape))
print("shape of school state (test) vector is {}".format(school_state_tst.shape))
```

```python

#teacher_prefix
categ_model.fit(x_train['teacher_prefix'].values)
teacher_prefix_tr = categ_model.transform(x_train['teacher_prefix'].values)
teacher_prefix_tst = categ_model.transform(x_test['teacher_prefix'].values)

print("-"*50)
print("shape of teacher_prefix (train) vector is {}".format(teacher_prefix_tr.shape))
print("shape of teacher_prefix (test) vector is {}".format(teacher_prefix_tst.shape))
```

```python
#project_grade_category
categ_model.fit(x_train['project_grade_category'].values)
project_grade_category_tr = categ_model.transform(x_train['project_grade_category'].values)
project_grade_category_tst = categ_model.transform(x_test['project_grade_category'].values)

print("-"*50)
print("shape of project_grade_category (train) vector is {}".format(project_grade_category_tr.shape))
print("shape of project_grade_category (test) vector is {}".format(project_grade_category_tst.shape))
```

```python
#encoding numerical feature
#price
from sklearn.preprocessing import Normalizer

num_model = Normalizer()
price_tr = num_model.fit_transform(x_train['price'].values.reshape(-1,1))
price_tst = num_model.fit_transform(x_test['price'].values.reshape(-1,1))

print("-"*50)
print("shape of price (train) vector is {}".format(price_tr.shape))
print("shape of price (test) vector is {}".format(price_tst.shape))
```

```python
#getting sentiment features
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
def getSentimentIntensity(corpus):
  neg = []
  neu = []
  pos = []
  comp = []

  sid = SentimentIntensityAnalyzer()
  for i in corpus:
    ss = sid.polarity_scores(i)
    neg.append(ss['neg'])
    neu.append(ss['neu'])
    pos.append(ss['pos'])
    comp.append(ss['compound'])

  return neg,neu,pos,comp
```

```python
essay_neg_tr,essay_neu_tr,essay_pos_tr,essay_comp_tr = getSentimentIntensity(x_train['essay'].values)
```

```python
essay_neg_tst,essay_neu_tst,essay_pos_tst,essay_comp_tst = getSentimentIntensity(x_test['essay'].values)
```

```python
#tfidf features
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_model = TfidfVectorizer()
tfidf_model.fit(x_train['essay'].values)
essay_tf_tr = tfidf_model.transform(x_train['essay'].values)
essay_tf_tst = tfidf_model.transform(x_test['essay'].values)

print("-"*50)
print("shape of essay tfidf (train) vector is {}".format(essay_tf_tr.shape))
print("shape of essay tfidf (test) vector is {}".format(essay_tf_tst.shape))
```

```python
import pickle
```

```python
#tfidf word2vec features
with open('/content/drive/My Drive/Colab Notebooks/Assignment 9/glove_vectors', 'rb') as f:
    model = pickle.load(f)
    glove_words =  set(model.keys())
```

```python
tfidf_model = TfidfVectorizer()
tfidf_model.fit(x_train['essay'].values)
dictionary = dict(zip(tfidf_model.get_feature_names(), list(tfidf_model.idf_)))
tfidf_words = set(tfidf_model.get_feature_names())
```

```python
from tqdm import tqdm
import numpy as np

essay_tfidf_w2v_tr = [] 
for sentence in tqdm(x_train['essay'].values):
    vector = np.zeros(300)
    tf_idf_weight =0
    for word in sentence.split():
        if (word in glove_words) and (word in tfidf_words):
            vec = model[word] 
            tf_idf = dictionary[word]*(sentence.count(word)/len(sentence.split()))
            vector += (vec * tf_idf)
            tf_idf_weight += tf_idf
    if tf_idf_weight != 0:
        vector /= tf_idf_weight
    essay_tfidf_w2v_tr.append(vector)

print('\n',len(essay_tfidf_w2v_tr))
print(len(essay_tfidf_w2v_tr[0]))
```

```python
essay_tfidf_w2v_tst = [] 
for sentence in tqdm(x_test['essay'].values):
    vector = np.zeros(300)
    tf_idf_weight =0
    for word in sentence.split():
        if (word in glove_words) and (word in tfidf_words):
            vec = model[word] 
            tf_idf = dictionary[word]*(sentence.count(word)/len(sentence.split()))
            vector += (vec * tf_idf)
            tf_idf_weight += tf_idf
    if tf_idf_weight != 0:
        vector /= tf_idf_weight
    essay_tfidf_w2v_tst.append(vector)

print('\n',len(essay_tfidf_w2v_tst))
print(len(essay_tfidf_w2v_tst[0]))
```

**Set - I (TFIDF)**

```python
#stacking all the features vectors to feed to the model
from scipy.sparse import hstack
import numpy as np

x_tr_tfidf = hstack((school_state_tr,teacher_prefix_tr,project_grade_category_tr,price_tr,essay_tf_tr,np.array(essay_neg_tr).reshape(-1,1),np.array(essay_neu_tr).reshape(-1,1),np.array(essay_pos_tr).reshape(-1,1),np.array(essay_comp_tr).reshape(-1,1))).tocsr()
x_tst_tfidf = hstack((school_state_tst,teacher_prefix_tst,project_grade_category_tst,price_tst,essay_tf_tst,np.array(essay_neg_tst).reshape(-1,1),np.array(essay_neu_tst).reshape(-1,1),np.array(essay_pos_tst).reshape(-1,1),np.array(essay_comp_tst).reshape(-1,1))).tocsr()

print('Final Features:')
print("-"*50)
print("shape of train vector is {}".format(x_tr_tfidf.shape))
print("shape of test vector is {}".format(x_tst_tfidf.shape))
```

**Set - II (TFIDF W2V)**

```python

x_tr_tfidf_w2v = hstack((school_state_tr,teacher_prefix_tr,project_grade_category_tr,price_tr,essay_tfidf_w2v_tr,np.array(essay_neg_tr).reshape(-1,1),np.array(essay_neu_tr).reshape(-1,1),np.array(essay_pos_tr).reshape(-1,1),np.array(essay_comp_tr).reshape(-1,1))).tocsr()
x_tst_tfidf_w2v = hstack((school_state_tst,teacher_prefix_tst,project_grade_category_tst,price_tst,essay_tfidf_w2v_tst,np.array(essay_neg_tst).reshape(-1,1),np.array(essay_neu_tst).reshape(-1,1),np.array(essay_pos_tst).reshape(-1,1),np.array(essay_comp_tst).reshape(-1,1))).tocsr()

print('Final Features:')
print("-"*50)
print("shape of train vector is {}".format(x_tr_tfidf_w2v.shape))
print("shape of test vector is {}".format(x_tst_tfidf_w2v.shape))
```
